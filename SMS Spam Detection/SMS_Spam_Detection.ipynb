```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
# Loading the dataset
file_path = 'spam.csv'
df = pd.read_csv(file_path)

print(df.head())
```

```python
# Map labels to numeric values
df['spamORham'] = df['spamORham'].map({'ham': 0, 'spam': 1})
```

```python
print(df.head())
```

```python
# Splitting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df['Message'], df['spamORham'], test_size=0.2, random_state=1)
```

```python
# Instantiate the TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer()

# Fit and transform the training data, and transform the testing data
training_data_tfidf = tfidf_vectorizer.fit_transform(X_train)
testing_data_tfidf = tfidf_vectorizer.transform(X_test)
```

```python
# Instantiate the Naive Bayes classifier
naive_bayes = MultinomialNB()

# Train the Naive Bayes model
naive_bayes.fit(training_data_tfidf, y_train)
```

```python
predictions = naive_bayes.predict(testing_data_tfidf)
```

```python
# Model Evaluation
print('Accuracy score: ', format(accuracy_score(y_test, predictions)))
print('Precision score: ', format(precision_score(y_test, predictions)))
print('Recall score: ', format(recall_score(y_test, predictions)))
print('F1 score: ', format(f1_score(y_test, predictions)))
```
